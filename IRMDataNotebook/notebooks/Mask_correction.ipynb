# Mask_correction

In this notebook, the Napari image viewer was used to correct the output Masks of the StarDist model. This script was also used to draw the cell masks and the skeletons of each cell mask. 

```python
import napari
from glob import glob
from skimage import io
from pathlib import Path
import numpy as np
from magicgui import magicgui
from napari.layers import Image, Labels
import skimage
from skimage.morphology import skeletonize
```

```python
masks_paths = glob(fr'..\A5\240325\StarDistOutput_corrected\*.tif')
img_paths = glob(fr'..\A5\240325\*.tif')
```

```python
viewer = napari.Viewer()
```

```python
i=10

img = io.imread(img_paths[i])
mask = io.imread(masks_paths[i])
 
viewer.add_image(img, scale = (3,1,1))
viewer.add_labels(mask, scale = (3,1,1))

# Create skeleton mask based on cell mas
#skeleton = medial_axis(mask,return_distance=False)
#viewer.add_labels(skeleton.astype(int), name='skeleton')
print(Path(img_paths[i]).stem)
```
